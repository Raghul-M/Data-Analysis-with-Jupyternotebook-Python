```python
import pandas as pd
```

## Joins

- Helps us in joining two data frames

##### Horizontal Joining ( Merge )

- Inner Join in Pandas
- Full Join in Pandas
- Left Join in Pandas
- Right Join in Pandas

#### Vertical Joining ( Concat )

```python
product=pd.DataFrame({
    'Product_ID':[101,102,103,104,105,106,107],
    'Product_name':['Watch','Bag','Shoes','Smartphone','Books','Oil','Laptop'],
    'Category':['Fashion','Fashion','Fashion','Electronics','Study','Grocery','Electronics'],
    'Price':[299.0,1350.50,2999.0,14999.0,145.0,110.0,79999.0],
    'Seller_City':['Delhi','Mumbai','Chennai','Kolkata','Delhi','Chennai','Bengalore']
})
```

```python
product.shape
```

```python
product
```

```python
customer=pd.DataFrame({
    'id':[1,2,3,4,5,6,7,8,9],
    'name':['Olivia','Aditya','Cory','Isabell','Dominic','Tyler','Samuel','Daniel','Jeremy'],
    'age':[20,25,15,10,30,65,35,18,23],
    'Product_ID':[101,0,106,0,103,104,0,0,107],
    'Purchased_Product':['Watch','NA','Oil','NA','Shoes','Smartphone','NA','NA','Laptop'],
    'City':['Mumbai','Delhi','Bangalore','Chennai','Chennai','Delhi','Kolkata','Delhi','Mumbai']
})
```

```python
customer.shape
```

```python
customer
```

### Inner Join

```python
### Joining based on one column

pd.merge(product,customer,on='Product_ID' , how = 'inner')
```

```python
### Joining based on two columns

pd.merge(product,customer,left_on='Product_name',right_on='Purchased_Product' , how = 'inner')
```

### Left Join

```python
### Joining based on one column

pd.merge(product,customer,on='Product_ID' , how = 'left')
```

```python
### Joining based on two columns

pd.merge(product,customer,left_on='Product_name',right_on='Purchased_Product' , how = 'left')
```

### Right Join


```python
### Joining based on one column

pd.merge(product,customer,on='Product_ID' , how = 'right')
```

```python
### Joining based on two columns

pd.merge(product,customer,left_on='Product_name',right_on='Purchased_Product' , how = 'right')
```

# Outer Join

```python
### Joining based on one column

pd.merge(product,customer,on='Product_ID' , how = 'outer')
```

```python
### Joining based on two columns

pd.merge(product,customer,left_on='Product_name',right_on='Purchased_Product' , how = 'outer')
```

# concat

```python
product_2020=pd.DataFrame({
    'Product_ID':[101,102,103],
    'Product_name':['Watch','Bag','Shoes'],
    'Category':['Fashion','Fashion','Fashion'],
    'Price':[299.0,1350.50,2999.0],
    'Seller_City':['Delhi','Mumbai','Chennai']
})
product_2020['Year'] = 2020
```

```python
product_2021=pd.DataFrame({
    'Product_ID':[104,105,106,107],
    'Product_name':['Smartphone','Books','Oil','Laptop'],
    'Category':['Electronics','Study','Grocery','Electronics'],
    'Price':[14999.0,145.0,110.0,79999.0],
    'Seller_City':['Kolkata','Delhi','Chennai','Bengalore']
})
product_2021['Year'] = 2021
```

```python
product_2020
```

```python
product_2021
```

```python
product_all = pd.concat( [ product_2020 , product_2021 ] , ignore_index = True  )
```

```python
product_all
```
