<b>Regex filtering</b>

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')] # Not contains
```

```python
import re
df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]
```

```python
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)] #For any words whether capital or small
```

```python
df.loc[df['Name'].str.contains('^pi[a=z]*',flags=re.I,regex=True)]
```
